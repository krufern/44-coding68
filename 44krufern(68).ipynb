```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# สร้างช่องกรอกน้ำหนักและส่วนสูง
weight_input = widgets.FloatText(description="น้ำหนัก (กก.):", style={'description_width': 'initial'})
height_input = widgets.FloatText(description="ส่วนสูง (ม.):", style={'description_width': 'initial'})

# ปุ่มคำนวณ
button = widgets.Button(description="คำนวณ BMI", button_style='success')

# แสดงผล
output = widgets.Output()

# ฟังก์ชันเมื่อคลิกปุ่ม
def calculate_bmi(b):
    with output:
        clear_output()
        try:
            weight = weight_input.value
            height = height_input.value
            bmi = weight / (height ** 2)

            if bmi < 18.5:
                result = f"ค่า BMI ของคุณ คือ {bmi:.2f}\nแปลผล: น้ำหนักน้อย\nคำแนะนำ: ทานอาหารให้ครบ 5 หมู่"
            elif bmi < 23.0:
                result = f"ค่า BMI ของคุณ คือ {bmi:.2f}\nแปลผล: น้ำหนักปกติ\nคำแนะนำ: รักษาสุขภาพนะคะ"
            else:
                result = f"ค่า BMI ของคุณ คือ {bmi:.2f}\nแปลผล: น้ำหนักเกิน\nคำแนะนำ: มาออกกำลังกายกันเถอะ"
            print(result)
        except:
            print("กรุณากรอกข้อมูลให้ถูกต้อง")

# เชื่อมปุ่มกับฟังก์ชัน
button.on_click(calculate_bmi)

# แสดง widget ทั้งหมด
display(weight_input, height_input, button, output)
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# สร้างช่องกรอกน้ำหนักและส่วนสูง
weight = widgets.FloatText(description='น้ำหนัก (กก.)', value=0.0)
height = widgets.FloatText(description='ส่วนสูง (ม.)', value=0.0)
button = widgets.Button(description="คำนวณ BMI", button_style='success')
output = widgets.Output()

# ฟังก์ชันเมื่อคลิกปุ่ม
def calculate_bmi(b):
    with output:
        clear_output()
        try:
            w = weight.value
            h = height.value
            if h <= 0:
                print("กรุณาใส่ส่วนสูงให้ถูกต้อง")
                return
            bmi = w / (h ** 2)
            print(f"ค่า BMI ของคุณ คือ {bmi:.2f}")
            if bmi < 18.5:
                print("แปลผล: น้ำหนักน้อย")
                print("คำแนะนำ: ทานอาหารให้ครบ 5 หมู่")
            elif bmi < 23.0:
                print("แปลผล: น้ำหนักปกติ")
                print("คำแนะนำ: รักษาสุขภาพนะคะ")
            else:
                print("แปลผล: น้ำหนักเกิน")
                print("คำแนะนำ: มาออกกำลังกายกันเถอะ")
        except:
            print("กรุณากรอกตัวเลขให้ถูกต้อง")

# เชื่อมปุ่มกับฟังก์ชัน
button.on_click(calculate_bmi)

# แสดง widget
display(widgets.VBox([weight, height, button, output]))
```
